```python
import numpy as np
import pandas as pd
```

```python
aData = np.random.uniform(10000, size= 100)
sData = pd.Series(aData)
```

```python
#m = resamples
#B = bootstrap trials
def se(m, B):
    boot = pd.DataFrame(index = [i+1 for i in range(10)])
    for i in range(B):
        temp = sData.sample(m)
        temp.index = [i+1 for i in range(m)]
        boot = pd.concat([boot, temp], axis = 1)
    boot.columns = [f'sample_{i+1}' for i in range(B)]
    theta = boot.max(axis= 0)
    return np.sqrt(theta.var())
```

```python
import matplotlib.pyplot as plt
```

```python
seplot = [se(m, 100) for m in range(1, 101)]
```

```python
plt.plot(seplot)
```

```python
seplot = [se(m, 200) for m in range(1, 101)]
```

```python
plt.plot(seplot)
```
